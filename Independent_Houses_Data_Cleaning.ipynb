```python
houses=houses.drop_duplicates()
```

```python
houses.shape
```

```python
houses.sample(2)
```

```python
houses.drop(['link','property_id'],axis=1,inplace=True)
```

```python
houses.rename(columns={'rate':'price_per_sqft','noOfFloor':'floorNum'},inplace=True)
```

```python
houses.shape
```

```python
houses['society'].value_counts().shape
```

```python
houses['society'].value_counts()
```

```python
houses['society']=houses['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★?','',str(name).strip())).str.lower()
```

```python
houses['society'].value_counts().shape
```

```python
houses['price']
```

```python
houses[houses['price']=='Price on Request'].shape
```

```python
houses.drop(houses[houses['price']=='Price on Request'].index,axis=0,inplace=True)
```

```python
houses.shape
```

```python
houses['price'].isna().sum()
```

```python
houses['price'].str.split().apply(len).value_counts()
```

```python
houses['price'].str.split().apply(lambda x: x[1]).unique()
```

```python
def treat_price(x):
  if x[-1]=='Lac':
    return round(float(x[0])/100,2)
  else :
    return round(float(x[0]),2)
```

```python
houses['price']=houses['price'].str.split().apply(treat_price)
```

```python
houses['price_per_sqft']
```

```python
houses['price_per_sqft']=houses['price_per_sqft'].apply(lambda ppsf: str(ppsf).strip().replace(' ','').replace('₹','').replace(',','').replace('/sq.ft.','')).astype('int64')
```

```python
houses.head(5)
```

```python
houses[['bedRoom','bathroom','balcony','additionalRoom']].isna().sum()
```

```python
houses.bedRoom.value_counts()
```

```python
houses['bedRoom']=houses['bedRoom'].str.split().str.get(0).astype('int64')
```

```python
houses['bathroom']=houses['bathroom'].str.split().str.get(0).astype('int64')
```

```python
houses['balcony'].str.split().str.get(0).value_counts()
```

```python
houses['balcony']=houses['balcony'].str.replace('No','0').str.split().str.get(0)
```

```python
houses['additionalRoom'].value_counts()
```

```python
houses['additionalRoom'].value_counts().shape
```

```python
houses['additionalRoom'].isna().sum()
```

```python
def handle(x):
  if x[0] == '0':
    return 0
  else  :
    return len(x)
```

```python
houses['additionalRoom']=houses['additionalRoom'].fillna('0').str.split(',').apply(handle)
```

```python
houses['floorNum']=houses['floorNum'].str.split().str.get(0).astype('float64')
```

```python
houses['facing'].fillna('NA',inplace=True)
```

```python
houses['area'] = round((houses['price']*10000000)/houses['price_per_sqft'])
```
