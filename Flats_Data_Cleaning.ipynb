```python
# import librariies
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings("ignore")
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
flats=pd.read_csv('https://raw.githubusercontent.com/Hercules2310/Real_estate_price_prediction/main/flats%20-%20flats.csv')
```

```python
flats.columns
```

```python
flats.shape
```

```python
flats.info()
```

```python
flats.duplicated().sum()
```

```python
flats.isna().sum()
```

```python
flats.drop(['link','property_id'],axis=1,inplace=True)
```

```python
flats.rename(columns={'area':'price_per_sqft'},inplace=True)
```

```python
flats.shape
```

```python
flats['society'].value_counts().shape
```

```python
flats['society'].value_counts()
```

```python
flats['society']=flats['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★?','',str(name).strip())).str.lower()
```

```python
flats['society'].value_counts().shape
```

```python
flats['price']
```

```python
flats[flats['price']=='Price on Request'].shape
```

```python
flats.drop(flats[flats['price']=='Price on Request'].index,axis=0,inplace=True)
```

```python
flats.shape
```

```python
flats['price'].isna().sum()
```

```python
flats['price'].str.split().apply(len).value_counts()
```

```python
flats['price'].str.split().apply(lambda x: x[1]).unique()
```

```python
def treat_price(x):
  if x[-1]=='Lac':
    return round(float(x[0])/100,2)
  else :
    return round(float(x[0]),2)
```

```python
flats['price']=flats['price'].str.split().apply(treat_price)
```

```python
flats['price_per_sqft']
```

```python
flats['price_per_sqft'].apply(lambda ppsf: str(ppsf).strip().replace(' ','').replace('₹','').replace(',','').replace('/sq.ft.','')).astype('int64')
```

```python
flats['price_per_sqft']=flats['price_per_sqft'].apply(lambda ppsf: str(ppsf).strip().replace(' ', '') \
                     .replace('₹', '').replace(',', '').replace('/sq.ft.', '')).astype('int64')
```

```python
flats.sample(5)
```

```python
flats[['bedRoom','bathroom','balcony','additionalRoom']].isna().sum()
```

```python
flats.bedRoom.value_counts()
```

```python
flats['bedRoom']=flats['bedRoom'].str.split().str.get(0).astype('int64')
```

```python
flats['bathroom']=flats['bathroom'].str.split().str.get(0).astype('int64')
```

```python
flats['balcony'].str.split().str.get(0).value_counts()
```

```python
flats['balcony']=flats['balcony'].str.replace('No','0').str.split().str.get(0)
```

```python
flats['additionalRoom'].value_counts()
```

```python
flats['additionalRoom'].value_counts().shape
```

```python
flats['additionalRoom'].isna().sum()
```

```python
def handle(x):
  if x[0] == '0':
    return 0
  else  :
    return len(x)
```

```python
flats['additionalRoom']=flats['additionalRoom'].fillna('0').str.split(',').apply(handle)
```

```python
flats['additionalRoom']
```

```python
flats['additionalRoom'].value_counts()
```

```python
flats[flats['balcony']=='3+'].sort_values(['bedRoom'])[['bedRoom','balcony']]
```

```python
flats['floorNum']
```

```python
flats['floorNum'].str.split().str.get(0).value_counts()
```

```python
flats['floorNum']=flats['floorNum'].str.split().str.get(0).replace('Ground',0).replace('Lower',0).replace('Basement',-1).str.extract(r'(\d+)').astype('float64')
```

```python
flats[flats['floorNum'].isna()] # info available in address
```

```python
flats.iloc[179,10]=6
```

```python
flats['facing'].value_counts()
```

```python
flats['facing'].isna().sum()
```

```python
flats['facing'].fillna('NA',inplace=True)
```

```python
flats.insert(loc=4,column='area',value=round((flats['price']*10000000)/flats['price_per_sqft']))
```

```python
flats.insert(loc=1,column='property_type',value='flat')
```
